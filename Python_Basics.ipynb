```python
#6. Write a function to demonstrate the behavior of mutable and immutable arguments.

#Mutable Argument
original_list = [1, 2, 3]
print(f"Outside function (before call): {original_list}")

def modify_list(my_list):
    print(f"Inside function (before change): {my_list}")
    my_list.append(4)
    print(f"Inside function (after change): {my_list}")

modify_list(original_list)
print(f"Outside function (after call): {original_list}")

#Immutable Argument
original_int = 10
print(f"Outside function (before call): {original_int}")
def modify_integer(my_int):
    print(f"Inside function (before change): {my_int}")
    my_int = my_int + 1
    print(f"Inside function (after change): {my_int}")

modify_integer(original_int)
print(f"Outside function (after call): {original_int}")
```

```python
#8. Write a program to demonstrate the use of logical operators.

age = int(input("Enter your age: "))
has_id = input("Do you have a id? (yes/no): ")

if age >= 18 and has_id == "yes":
    print("You can enter the club.")
else:
    print("You cannot enter the club.")

# Using 'or' operator
if age < 18 or has_id == "no":
    print("Either you are too young or you don't have a id.")

# Using 'not' operator
if not has_id == "yes":
    print("You need to bring your id.")
```

```python
#13. Write a Python program to find the sum of all even numbers between 1 and 50.

even_sum = 0
for i in range(1,51):
  if i % 2 == 0:
    even_sum += i
print(f"Sum of even numbers from 1 to 50 is {even_sum}")
```

```python
#14.. Write a program to reverse a string using a while loop.

original_string = input("Enter a string to reverse: ")

reversed_string = ""

index = len(original_string) - 1

while index >= 0:
    reversed_string = reversed_string + original_string[index]
    index = index - 1

print(f"Original: {original_string}")
print(f"Reversed: {reversed_string}")
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"The factorial of {num} is: {factorial}")
```
